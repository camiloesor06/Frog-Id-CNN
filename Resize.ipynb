# Notebook for resize images prior to labeling for CNN Training

```python
from google.colab import files

# Upload images
uploaded = files.upload()
```

```python
import os
from PIL import Image
```

```python
# create output folder for resized pictures
output_folder = 'redimensionadas'
os.makedirs(output_folder, exist_ok=True)

# size
target_size = (416, 416)  # Altura y ancho


for filename in uploaded.keys():
    img = Image.open(filename).convert('RGB')  # Asegura 3 canales (RGB)


    img_resized = img.resize(target_size, Image.Resampling.LANCZOS)


    img_resized.save(os.path.join(output_folder, filename))
    print(f'{filename} redimensionada a {target_size}')
```

```python
from google.colab import files
import shutil
```

```python
# Compress to zip file
shutil.make_archive('Pristimantis palmeri redim', 'zip', output_folder)

# Download
files.download('Pristimantis palmeri redim.zip')
```
